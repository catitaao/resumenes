## Introducción a los S.I.

### Contexto

En un negocio existe una importante acumulación de datos, de los cuales muchos se pierden, no todos son relevantes, no todos son correctos y no todo es útil. Para solucionar esto, es necesario fijar un espacio a cubrir, clasificar, entender lo que se está generando y luego producir la información necesaria y correcta.

Además, los proyectos (y los sistemas) se deben alinear a la **misión** de la organización. Para esto, cada actividad debe agregar valor para aportar en el cumplimiento de la misión.

- **Misión:** Es la razón de la existencia, el objetivo y el para qué fue creada la empresa.
- **Visión:** Es la proyección a futuro de la empresa, su meta.
- **Estrategia:** Es la dirección y el conjunto de decisiones que permite cumplir la misión y acercar a la visión. Esto consiste en un conjunto de principios, objetivos y declaraciones que permiten tomar decisiones.

En este curso se busca encontrar maneras de apoyar la toma de decisiones, agregando valor a la organización. Para esto, es fundamental:
1. (Re)conocer la situación y su entorno para así entender el desafío.
2. Recopilar información adecuada.
3. Almacenar la información apropiada y seguramente.
4. Analizar la información de manera objetiva y agnóstica.
5. Categorizar la información como estratégica, de gestión u operativa.
6. Alinear los resultados de la decisión con la misión y visión de la organización. 
7. Obtener los resultados.
8. Iterar hasta lograr los objetivos.

**Para esto, los S.I. permiten cumplir objetivos de negocio estratégicos, como:**

- **Excelencia operacional (Bajar costos):** Al utilizar S.I., mejoran la eficiencia y productividad, bajando costos.

- **Nuevos productos, servicios y modelos de negocio (Vender (algo) más a los mismos):** Esto se logra mediante modelos de negocios, los cuales describen la forma en la que la compañía produce y entrega productos o servicios para crear valor para clientes y utilidad para la empresa.

- **Vínculo con clientes y proveedores (Retener):** Al tener una buena relación con los clientes, se logra que estos regresen. Aumentando ingresos y rentabilidad (**C**ustomer **R**elationship **M**anagement). Mientras que con los proveedores, permite tener información importante y disminuye costos.

- **Mejorar la toma de decisiones (Predecir la siguiente compra):** Mediante el uso de S.I, se mejora la precisión al momento de predecir decisiones de la empresa. Disminuyendo casos de sobre o sub producción, largos tiempos de respuestas y mal aprovechamiento de recursos.

- **Supervivencia/Resiliencia (Adaptarse):** Los sistemas de información entregan información escencial para la empresa, manteniendola al día con los cambios en la industria y mantenerse vigente a la normativa.

- **Obtener ventaja competitiva**: Ofreciendo algo único, distinto a la competencia, más calidad y menos costo, comunicación constante con clientes y proveedores.

La empresa y los sistemas de información son dependientes. Ya que la empresa trbaja en objetivos de negocios estratégicos y procesos de negocios, los cuales obtiene a través de un software. El cual, a su vez requiere de hardware, administración de datos y telecomunicaciones.

### ¿Qué es un sistema de información?

Un sistema es un grupo interdependiente de entidades, las cuales se caracterizan por atributos los cuales se relacionan entre sí mediante actividades, conformando un conjunto con un objetivo específico. Este sistema presenta fronteras delimitadas y también interactúa con su entorno.

Al aplicarlo en un S.I., este recopila datos, los almacena (temporal o permanente), los procesa mediante reglas de negocios y retorna información. Esta última sirve para el proceso de toma de decisiones.

datos -> recopilación -> almacenamiento -> preservación -> resultados
(input)    (proceso)      (subsistema)       (proceso)    (subsistema)

Los programas computacionales son la base técnica de un sistema de información.

##### **Composición de un S.I.**
- **Procesos de negocios:** Pasos diseñados para obtener un objetivo.
- **Datos:** Fluyen a través del sistema para convertirse en información.
- **Software:** La lógica sobre la que opera un S.I.
- **Hardware:** Componentes físicos que permiten hacer uso del software.
- **Personas/organizaciones:** Quienes participan en los procesos, normalmente como usuarios.

👀 Los **datos** son hechos que se registran al sistema. A partir de esto, se retornan datos organizados de manera significativa, la **información**. Es importante que los datos entregados sean "buenos" para poder recibir información de calidad.

##### Información
Para asegurar la calidad de las decisiones, se deben conocer las características de la información:
- Finalidad: Debe presentar un propósito al momento de transmitirla.
- Modo y formato: La comunicación de información tiene un carácter sensorial, ya que los humanos reciben la información de manera verbal, visual o escrita.
- Redundancia/eficiencia: Se busca un lenguaje eficiente, solamente con la información necesaria para un nivel óptimo de transmisión de información. Cuando esto no se cumple, la información es excesiva y redundante.
- Velocidad: La velocidad de transmisión influye en el tiempo que tarda el receptor en comprenderla, se busca minimizar esos tiempos.
- Frecuencia: Es necesario encontrar un balance. La información excesiva produce ruidos/distracciones, mientras que la mínima puede no informar lo suficiente.
- Determinístico/probabilístico: La información puede presentar un solo valor resultante (determinística) o múltiples resultados acompañados de sus probabilidades (probabilístico).
- Costo: El costo puede limitar la obtención de información, por lo que se deben ir evaluando constantemente.
- Valor: Esto depende del modo, velocidad, frecuencia, características, confiabilidad y validez de la información.
- Confiabilidad y precisión: Estos elementos son caros, por lo que se debe balancear su relación con el costo.
- Exactitud: Esto representa una aproximación a que tan verdadera es la información.
- Validez: Mide el grado de representación de la información en cuanto a lo que quiere representar.
- Actualidad: La antigüedad de la información.

#### Estudio de un S.I.

Las empresas tienen un proceso llamado la **cadena de valor** en el cual procesan los datos adquiridos para convertirlos en información útil. Donde miden el valor de un sistema según su capacidad para generar mejores decisiones, aumentar la eficiencia y aumentar la rentabilidad.

A partir de esto, surgen distintos enfoques contemporáneos para el estudio de S.I.
- Técnico: Enfatiza modelos basados en matemáticas. Usan informática, investigación de operaciones y ciencia de la administración.
- Comportamiento: Ocupa el comportamiento que surge del desarrollo y mantenimiento a largo plazo de los sistemas de información. Donde información estratégica de negocios, diseño, implementación, utilización y administración no se pueden analizar útilmente con la metodología técnica.

#### Gestión de sistemas (sistemas sociotécnicos)

Los principales actores de los S.I. son:
- Empresas (que invierten y buscan obtener valor de la tecnología)
- Proveedores de hardware y software
- Gerentes y empleados
- Contexto legal, social y cultural

Mediante esta gestión, se busca optimizar el desempeño de la empresa. Es importante ponerle atención a componentes técnicos, actualizar y diseñar la tecnología para que se ajuste a las necesidades.

## Organizaciones y Sistemas de Información

### Procesos de negocios

![Screenshot 2024-10-04 at 1.12.13 PM.png](<attachment:Screenshot 2024-10-04 at 1.12.13 PM.png>)

Un **proceso** es una secuencia de actividades que se hacen para lograr el resultado que se busca. Por lo que cuentan con un propósito, un inicio, participantes (personas, sistemas, internos, externos), actividades con reglas entre si, secuencialidad y un término. Las organizaciones pueden ser conjuntos de procesos de y del negocio.

##### Cadena de procesos
Evento gatillador -> Proceso (compuesto de actividades) -> Resultado

Este proceso es cíclico, donde los resultados tienden a gatillar nuevos eventos. De esta forma, se crea una arquitectura de procesos. Los eventos gatilladores son necesidades (tanto internas como externas) o productos intermedios.
La red formada por interacción en distintos niveles entre estos procesos constituye una organización.
La **cadena de valor** representa, a nivel macro, los procesos operativos y su aporte a la misión. Le entrega a la empresa una vista de organización, actividades, procesos y el aporte y relación entre ellos y la empresa.

### S.I. en las distintas áreas de una organización

#### Ejemplos de S.I.

- **Supply Chain Management Systems (SCM)**: Gestión del flujo de bienes, datos y finanzas relacionados con un producto o servicio (desde la adquisición de materias primas hasta la entrega del producto en su destino).

- **Customer Relationship Management (CRM)**: Examinan a los clientes desde una perspectiva multifacética. Usan un conjunto integrado de aplicaciones para abordar todos los aspectos de la relación con los clientes (incluyendo servicios al cliente, ventas y marketing).

- **Enterprise Resource Planning (ERP)**: Sistema de software que incluye todas las herramientas y procesos necesarios para dirigir una empresa exitosa (incluidos RR. HH., manufactura, cadena de suministro, finanzas, contabilidad y más).

#### Categorías de S.I.

- **Según área/integración de áreas**: Sistemas empresariales (ERP), sistemas de administración de la cadena de suministro (SCM), sistemas de administración de relaciones con cliente (CRM).

- **Según nivel de jerarquía organizacional**: Sistema de procesamiento de transacciones (TPS), sistemas para inteligencia de negocios (BI), soporte gerencial (MIS, Management Information System), soporte de Decisiones (DSS, Decision Support System), soporte Ejecutivo (ESS, Executive Support System).

##### Tipos de S.I. 

- Procesamiento de transacciones (**TPS**, **T**ransaction **Processing** **S**ystem): Registran diariamente transacciones y eventos necesarios. Utilizado por jefes de operaciones y sus equipos, les permite medir metas predefinidas y concretas y monitorear el estado de la operación y/o su relación con el entrono. (ej: orden de venta, entregas, pago de sueldos, orden de compra...).

- Inteligencia de negocios (**BI**, **B**usiness **I**ntelligence):

    - Soporte gerencial (MIS, ***M**anagement **I**nformation **S**ystem*): Se trabaja en niveles medios (jerarquía), entregando resúmenes y reportes sobre la operación de la empresa, usa datos del TPS, también responde preguntas cotidianas de procedimientos definidos y posee poca capacidad analítica.

    - Soporte de decisiones (DSS, ***D**ecision-**S**upport **S**ytem*): Utilizado en gerencia media para el apoyo en toma de decisiones **no** rutinarias (ej: pronósticos de producción y adquisiciones), utiliza información tanto interna como externa de modelos (ej: estimación de viajes, determinación de precios)

    - Soporte *top-level executives* (ESS, ***E**xecutive **S**upport **S**ystem*): Utilizado por gerencia superior y directorio para la toma de decisiones **no** rutinarias que requieren juicio, evaluación y visión. Utiliza datos de eventos externos y de los resultados de **MIS** y **DSS** (ej: dashboards).

- Sistemas para integrar áreas:

    - Empresariales (ERP, ***E**nterprise **R**esource **P**lanning*)

    - Administración de cadena de suministro (SCM, ***S**upply **C**hain **M**gmt*)

    - Administración de relaciones con clientes (CRM, ***C**ustomer **R**elationship **M**gmt*)

    - Administración de conocimiento (KMS)

#### Métodos de mejora de procesos con S.I
- Implementación de una arquitectura de sistemas.
- Automatizar la eficiencia de procesos existentes (automatizando procesos manuales y disminuyendo errores humanos).
- Permitiendo nuevos procesos (cambiando el flujo de información, reemplazando actividades secuenciales por paralelas, eliminando retrasos en los procesos de decisión y apoyando nuevos modelos de negocios).

### Organizaciones

Una organización es una estructura de personas regulada por un conjunto de normas (**procesos**) en función de sus objetivos(**resultados -> misión**).

![Screenshot 2024-10-04 at 9.02.01 PM.png](<attachment:Screenshot 2024-10-04 at 9.02.01 PM.png>)

Las organizaciones son sistemas sociales con cultura, políticas, protocolos y conflictos (surgen y se resuelven). Son complejas y dinámicas, siendo afectadas por su entorno y organizandose mediante jerarquías, roles, funciones y responsabilidades. 

#### Efectos de los S.I.

- Impactos Económicos: Cambian los costos relativos del capital y
los costos de la información (optimizando), pueden ser factor de producción (sustituyendo capital/trabajo), afectan al costo y a la calidad de la información y cambian la economía de la información, ayudan a reducir costos de transacción, tienden a requerir de una menor cantidad de empleados y fácil expansión.

- Impactos organizacionales y de comportamiento: Mediante TI es posible aplanar la organización (llevando la toma de decisiones a niveles más bajos y requiriendo de menos gerentes). *Knowledge work* es el nombre que se le da al formato organizacional post-industrial donde la autoridad prioriza el conocimiento sobre cargos formales.
    - La gestión por algoritmos utiliza los datos para que los algoritmos puedan llevar a cabo funciones que normalmente harían los gestores.

- Resistencia al cambio: Al influir en el acceso a la información, se ligan a la política de la organización. Pueden cambiar la estructura, la política y el trabajo de una organización y permiten una buena **gestión del cambio**, evitando resistencia a este (gran causa de falla).

- Estrategias competitivas usando SI: Facilitan liderazgo del bajo costo, **diferenciación** del producto, enfoque en **nichos** de mercado y refuerzan la **relación con clientes y proveedores**. Sus desafíos son:
    - Mantener ventaja competitiva, ya que la competencia también puede aplicar las mismas estrategias.
    - Alinear las TI con los objetivos, esto se logra analizando la estructura de la industria y la cadena de valor de la empresa.
    - Gestionar transiciones estratégicas, con el fin de adoptar sistemas estratégicos.

### Estrategias

Una estrategia presenta una mirada general de la organización. Al aplicarlo a TI, se da un enfoque específico sobre problemas a resolver mediante tecnología.

El área de TI, en una organización, puede estar a diferentes niveles:
- Operativo: Netamente de **apoyo**, alejado del propósito de la organización.
- Gestión: Área de apoyo con **uso intensivo de información** para calcular indicadores, generar eficiencia organizacional y apoyar la eficacia de la gestión.
- Estratégico: Área **clave**, muchas veces el negocio es tecnológico.

El área TI puede entonces ser tratada como:
- Área de apoyo transversal: Por ejemplo, unidades de soporte.
- Socio tecnológico: En este caso satisface las necesidades de tecnología en la organización en forma sistemática. Hay una unidad de tecnología.
- Área de proyectos: Implica formar una oficina de proyectos tecnológicos y un conocimiento de la organización para generar soluciones transversales y coordinadas (PMO por medio).

#### Encontrar el problema

Existe la metodología informal **"QQCC"**, la cual busca encontrar las brechas, dónde se producen e investigar el Qué, Quién, Cómo y Cuándo. Con la misión de detectar necesidades. Siendo esta una versión informal de la ingeniería de requerimientos.
- ¿**Qué** es lo que se espera obtener? y ¿**Qué** es lo que se está obteniendo?
- ¿**Quién** (proceso, persona, plataforma) lo entrega?
- ¿**Cómo** se realiza? (el proceso)
- ¿**Cuándo** se debe entregar? y ¿**Cuándo** se debe mejorar?

## Modelamiento de Procesos

Es importante conocer el contexto de una organización (procesos y funciones, macroprocesos/procesos) al momento de desarrollar un SI. 
Un macroproceso se asocia a áreas grandes, como adquisiciones, producción, finanzas, etc. Mientras que los procesos tienden a ser más "cotidianos" como vender, comprar, contabilizar, etc.

Los procesos de negocios son "horizontales", ya que atraviesan diversas áreas ("horizontales") de un organigrama de la organización. Van debajo de los macroprocesos y los atraviesan.

##### Matriz **RACI**

Describe la asignación de responsabilidades en las actividades definidas por un proceso de negocios. Refiriendose a:
- **R**esponsible (responsable)
- **A**ccountable (Autoridad)
- **C**onsulted (Consultado)
- **I**nformed (informado)

Al momento de crearla, se crea una lista de tareas en un eje y una de cargos en el otro. Asignando una de las 4 responsabilidades (RACI) a cada cargo en cada tarea.
